```
from roboflow import Roboflow

rf = Roboflow(api_key="vs7yVxvN1BkvbGSZJwfC")
project = rf.workspace().project("turkish-number-plates")
version = project.version(1)

# Modeli indir
model = version.model
print("Model hazır:", project.name, "v" + str(version.version))
```

```
# Model ağırlıklarını YOLOv8 formatında indir
import os, shutil

version.download("yolov8", location="./turkish-plates-model")

# best.pt dosyasını bul ve kopyala
for root, dirs, files_list in os.walk("./turkish-plates-model"):
    for f in files_list:
        if f.endswith(".pt"):
            src = os.path.join(root, f)
            shutil.copy(src, "plate-detector.pt")
            print(f"Model bulundu ve kopyalandı: {src} → plate-detector.pt")
            print(f"Boyut: {os.path.getsize('plate-detector.pt') / (1024*1024):.1f} MB")
            break

# Eğer .pt yoksa data.yaml'ı göster (dataset olabilir)
if not os.path.exists("plate-detector.pt"):
    print("Hazır .pt bulunamadı. İndirilen dosyalar:")
    for root, dirs, files_list in os.walk("./turkish-plates-model"):
        for f in files_list:
            print(f"  {os.path.join(root, f)}")
```
